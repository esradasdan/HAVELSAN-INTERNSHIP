### Reading Files

### Preprocessing

```python
def calculate_RunId_X(df):
    
    if df['Load'] == 0:
        return df['RunId'] * 2 -1
    else:
        return df['RunId'] * 2
    
    
def resample(df):
        df = df.groupby(df.index //4).mean()
        return df
        

def preprocessing(df):
    
    # Getting only SeaTemp = 20 and Windspeed = 0
    df = df[(df.SeaTemp == 'SW20') & (df.WindSpeed == 'WC0')]
    
    # select features
    abr_sensor_list = ['E02005', 'E02006', 'E02056', 'E03760', 'G00027', 'G00108', 'G02011', 'N02015', 'P00023', 'P01005', 'P01302', 'P01303', 'P01600', 'P01602', 'P02055', 'P02065', 'P02066', 'P02071', 'P02072', 'Q02004', 'T00002', 'T01010', 'T01011', 'T01350', 'T01351', 'T01601', 'T01603', 'T02014', 'T02040', 'T02041', 'T02042', 'T02044', 'T04600', 'Z00518', 'Z01970', 'Z02013']
    ic = ['Load','RunId']
    ic.extend(abr_sensor_list)
    df = df[ic]
    
    # LoadType 0-1
    df['Load'] = df['Load'].apply(lambda x: 0 if x == 'FAL' else 1)
    
    # interpolate
    df = df.interpolate().fillna(df.mean())
    
    # create RunID_X column
    runId_column = df.apply(calculate_RunId_X, axis=1)
    df['RunID_X'] = runId_column
    df['RunID_X'] = df['RunID_X'].astype(int)
    
    # resampling
    df = df.groupby('RunID_X').apply(resample)
    
    # drop Load, RunId, RunID_X
    df = df.drop(['Load','RunId', 'RunID_X'], axis =1 )


    return df
```

```python
def train_test(df_train, df_test):
    
    # X_train - X_test - NOx_train - NOx_test
    NOx_train = df_train['Z01970'] 
    NOx_test = df_test['Z01970']

    X_train = df_train.drop(['Z01970'], axis = 1)
    X_test = df_test.drop(['Z01970'], axis = 1)
    
    return X_train, NOx_train, X_test, NOx_test
```

```python
def scale_data(X_train, NOx_train, X_test, NOx_test):
    
    x_train_scaler = MinMaxScaler()
    X_train_scaled = x_train_scaler.fit_transform(X_train)

    x_test_scaler = MinMaxScaler()
    X_test_scaled = x_test_scaler.fit_transform(X_test)

    NOx_train_scaler = MinMaxScaler()
    NOx_train_scaled = NOx_train_scaler.fit_transform(np.array(NOx_train).reshape(-1,1))

    NOx_test_scaler = MinMaxScaler()
    NOx_test_scaled = NOx_test_scaler.fit_transform(np.array(NOx_test).reshape(-1,1))
    
    return X_train_scaled, NOx_train_scaled, X_test_scaled, NOx_test_scaled, NOx_test_scaler
```

#### Get X_train, NOx_train, X_test, NOx_test for each Fault Class


```python
# M0000_X_train, M0000_NOx_train, M0000_X_test, M000_NOx_test

m0000_train = preprocessing(m0000_train)
m0000_test = preprocessing(m0000_test)

M0000_X_train, M0000_NOx_train, M0000_X_test, M0000_NOx_test = train_test(m0000_train, m0000_test)

M0000_X_train_scaled, M0000_NOx_train_scaled, M0000_X_test_scaled, M0000_NOx_test_scaled, M0000_NOx_test_scaler = scale_data( M0000_X_train, M0000_NOx_train, M0000_X_test, M0000_NOx_test)
    

# M2503_X_train, M2503_NOx_train, M2503_X_test, M2503_NOx_test

m2503_train = preprocessing(m2503_train)
m2503_test = preprocessing(m2503_test)

M2503_X_train, M2503_NOx_train, M2503_X_test, M2503_NOx_test = train_test(m2503_train, m2503_test)

M2503_X_train_scaled, M2503_NOx_train_scaled, M2503_X_test_scaled, M2503_NOx_test_scaled, M2503_NOx_test_scaler = scale_data( M2503_X_train, M2503_NOx_train, M2503_X_test, M2503_NOx_test)

    
# M2506_X_train, M2506_NOx_train, M2506_X_test, M2506_NOx_test

m2506_train = preprocessing(m2506_train)
m2506_test = preprocessing(m2506_test)

M2506_X_train, M2506_NOx_train, M2506_X_test, M2506_NOx_test = train_test(m2506_train, m2506_test)

M2506_X_train_scaled, M2506_NOx_train_scaled, M2506_X_test_scaled, M2506_NOx_test_scaled, M2506_NOx_test_scaler = scale_data( M2506_X_train, M2506_NOx_train, M2506_X_test, M2506_NOx_test)


# M2508_X_train, M2508_NOx_train, M2508_X_test, M2508_NOx_test

m2508_train = preprocessing(m2508_train)
m2508_test = preprocessing(m2508_test)

M2508_X_train, M2508_NOx_train, M2508_X_test, M2508_NOx_test = train_test(m2508_train, m2508_test)

M2508_X_train_scaled, M2508_NOx_train_scaled, M2508_X_test_scaled, M2508_NOx_test_scaled, M2508_NOx_test_scaler = scale_data( M2508_X_train, M2508_NOx_train, M2508_X_test, M2508_NOx_test)
```

### Predictions

```python
def predict_NOx(X_test_scaled, NOx_test_scaler, gpr):
    
    NOx_pred_scaled = gpr.predict(X_test_scaled, return_std=False)
    NOx_pred_scaled = NOx_pred_scaled.flatten()
    NOx_pred_scaled = NOx_pred_scaled.reshape(-1, 1)
    NOx_pred = NOx_test_scaler.inverse_transform(NOx_pred_scaled)
    
    return NOx_pred
```

```python
def fault_class_plot(runid, X_test, fault_class, load_type):
    
    
    runid_x = runid * 2 -1
    
    if load_type == "FAL":
    
        nox_fal = X_test.loc[runid_x]['Z01970']
        pred_m0000_model_fal = X_test.loc[runid_x]['pred_m0000']
        pred_m2503_model_fal = X_test.loc[runid_x]['pred_m2503']
        pred_m2506_model_fal = X_test.loc[runid_x]['pred_m2506']
        pred_m2508_model_fal = X_test.loc[runid_x]['pred_m2508']
        
        label_ = ""
        label_ += fault_class + "_FAL_test"
        

        plt.plot(nox_fal.values,label = label_ , color="red")  
        plt.plot(pred_m0000_model_fal.values ,label="m0000_model_FAL_pred", color="blue")
        plt.plot(pred_m2503_model_fal.values ,label="m2503_model_FAL_pred", color="green")
        plt.plot(pred_m2506_model_fal.values ,label="m2506_model_FAL_pred", color="pink")
        plt.plot(pred_m2508_model_fal.values ,label="m2508_model_FAL_pred", color="yellow")
        
        plt.legend(loc="upper right")

        plt.tight_layout()
        title_ = fault_class + " Dataset - NOx Predictions  |  "  +  "  RunId:" + str(runid) + "  | FAL"
        plt.title(title_)
        plt.show()
        
    elif load_type == "FAU":
        
    
        nox_fau = X_test.loc[runid_x + 1]['Z01970']
        pred_m0000_model_fau = X_test.loc[runid_x + 1]['pred_m0000']
        pred_m2503_model_fau = X_test.loc[runid_x + 1]['pred_m2503']
        pred_m2506_model_fau = X_test.loc[runid_x + 1]['pred_m2506']
        pred_m2508_model_fau = X_test.loc[runid_x + 1]['pred_m2508']

        label_ = ""
        label_ += fault_class + "_FAU_test"
        

        plt.plot(nox_fau.values,label = label_ , color="red")  
        plt.plot(pred_m0000_model_fau.values ,label="m0000_model_FAU_pred", color="blue")
        plt.plot(pred_m2503_model_fau.values ,label="m2503_model_FAU_pred", color="green")
        plt.plot(pred_m2506_model_fau.values ,label="m2506_model_FAU_pred", color="pink")
        plt.plot(pred_m2508_model_fau.values ,label="m2508_model_FAU_pred", color="yellow")
        
        plt.legend(loc="upper right")

        plt.tight_layout()
        title_ = fault_class + " Dataset - NOx Predictions  |  "  +  "  RunId:" + str(runid) + "  | FAU"
        plt.title(title_)
        plt.show()
```
